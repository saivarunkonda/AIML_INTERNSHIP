```python
df['Plot'] = df['Plot'].str.replace('[^ws]','', regex = False)
```

```python
df['Plot']
```

```python
# df['Key_words'] = ''
```

```python
df['Key_words'] = ""
r = Rake()

for index, row in df.iterrows():
    r.extract_keywords_from_text(row['Plot'])
    key_words_dict_scores = r.get_word_degrees()
    df.at[index, 'Key_words'] = list(key_words_dict_scores.keys())
```

```python
df['Key_words']
```

```python
df['Genre'] = df['Genre'].astype(str).map(lambda x: x.split(','))
df['Actors'] = df['Actors'].map(lambda x: x.split(',')[:3])
df['Director'] = df['Director'].map(lambda x: x.split(','))
```

```python
df['Director']
```

```python
for index, row in df.iterrows():
    df.at[index, 'Genre'] = [x.lower(). replace(' ', '')for x in row['Genre']]
    df.at[index, 'Actors'] = [x.lower(). replace(' ', '')for x in row['Actors']]
    df.at[index, 'Director'] = [x.lower(). replace(' ', '')for x in row['Director']]
```

```python
df['Actors']
```

```python
df['bag_of_words'] =  ''

columns = ['Genre','Actors','Director','Key_words']

for index, row in df.iterrows():
    words =''
    for col in columns:
        words += ' '.join(row[col]) + ' '
    df.at[index, 'bag_of_words'] = words

df['bag_of_words'] = df['bag_of_words'].str.strip().str.replace('   ', ' ').str.replace('  ', ' ')
```

```python
df2 = df[['Title', 'bag_of_words']]
```

```python
df2
```

```python
count = CountVectorizer()

count_matrix = count.fit_transform(df2['bag_of_words'])
```

```python
print(count_matrix)
```

```python
cosine_similarity = cosine_similarity(count_matrix, count_matrix)
```

```python
cosine_similarity.shape
```

```python
indicies = pd.Series(df2['Title'])
```

```python
indicies
```

```python
def recommanded(title, cosine_sim = cosine_similarity):
    recommandation = []
    idx = indicies[indicies ==title].index[0]
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending=False)
    print(score_series)
    top_5 = list(score_series[1:6].index)

    for i in top_5:
        recommandation.append(list(df['Title'])[i])
    return recommandation
    
```

```python
print(recommanded('On the Waterfront'))
```
