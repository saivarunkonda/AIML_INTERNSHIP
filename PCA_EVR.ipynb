```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Step 1: Load the dataset (Iris dataset as an example)
data = load_iris()
X = data.data
y = data.target

# Step 2: Standardize the data
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# Step 3: Compute the Covariance Matrix
cov_matrix = np.cov(X_std, rowvar=False)

# Step 4: Compute Eigenvectors and Eigenvalues
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Step 5: Calculate the Explained Variance Ratios (EVR)
explained_variance_ratio = eigenvalues / sum(eigenvalues)

# Step 6: Determine the Number of Principal Components
cumulative_variance = np.cumsum(explained_variance_ratio)
num_components = np.argmax(cumulative_variance >= 0.95) + 1  # Keep components explaining at least 95% of variance

# Step 7: Project Data onto New Space
pca = PCA(n_components=num_components)
X_pca = pca.fit_transform(X_std)

# Step 8: Visualize the Results
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap=plt.cm.Set1, edgecolor='k')
plt.title('PCA of Iris Dataset')
plt.xlabel(f'Principal Component 1 (EVR={explained_variance_ratio[0]:.2f})')
plt.ylabel(f'Principal Component 2 (EVR={explained_variance_ratio[1]:.2f})')
plt.show()
```
