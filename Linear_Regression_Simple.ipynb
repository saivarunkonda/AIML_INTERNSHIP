```python
import  numpy  as  np
import  matplotlib.pyplot  as  plt
from  sklearn.linear_model  import  LinearRegression
```

```python
#  Let's  assume  these  are  the  ages  and  weights  of  different  people

ages  =  np.array([20,  25,  30,  35,  40,  45,  50,  55,  60,  65,70]).reshape(-1,1)
weights  =  np.array([70,  72,  75,  80,  85,  88,  92,  95,  97,  100,110]).reshape(-1,1)
```

```python
ages
```

```python
model = LinearRegression()
```

```python
model.fit(ages,weights)
```

```python
# #  Now  we  can  get  the  slope  (m)  and  y-intercept  (c)  of  the  line
m  =  model.coef_
c  =  model.intercept_
```

```python
print(m,c)
```

```python
predicted_weigts = model.predict(ages)
```

```python
predicted_weigts
```

```python
weights
```

```python
model.score(ages,predicted_weigts)
```

```python
model.score(ages,weights)
```

```python
#  Plot  the  actual  data  points  (in  blue)  and  the  line  of  best  fit  (in  red)
plt.scatter(ages,  weights,  color='blue')
plt.plot(ages,  predicted_weigts,  color='red')
plt.xlabel('Age  (years)')
plt.ylabel('Weight  (kg)')
plt.title('Linear  Regression  of  Weight  on  Age')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['Electronics', 'Clothing', 'Home Goods', 'Beauty Products']
average_inventory = [750, 1200, 500, 300]
target_inventory = [500, 1500, 300, 250]

# Plotting
fig, ax = plt.subplots()
bar_width = 0.35
index = range(len(categories))

bar1 = ax.bar(index, average_inventory, bar_width, label='Average Inventory')
bar2 = ax.bar([i + bar_width for i in index], target_inventory, bar_width, label='Target Inventory')

# Adding labels
ax.set_xlabel('Product Category')
ax.set_ylabel('Inventory Level')
ax.set_title('Comparison of Average and Target Inventory Levels by Product Category')
ax.set_xticks([i + bar_width/2 for i in index])
ax.set_xticklabels(categories)
ax.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
years = ['\'90', '\'91', '\'92', '\'93', '\'94', '\'95', '\'96', '\'97']
deals_in_china = [19, 548, 125, 68, 1661, 767, 2592, 1456]
deals_outside_china = [150, 397, 159, 1273, 265, 962, 969, 6426]

# Stacked Bar Chart
fig, ax = plt.subplots()
bar_width = 0.5

# Bottom bar (deals outside China)
bars_bottom = ax.bar(years, deals_outside_china, bar_width, label='Deals Outside China', color='lightblue')

# Top bar (deals in China)
bars_top = ax.bar(years, deals_in_china, bar_width, bottom=deals_outside_china, label='Deals in China', color='darkblue')

# Adding labels and legend
ax.set_xlabel('Year')
ax.set_ylabel('Deals in Millions ($)')
ax.set_title('Chinese Acquisitions: Inside & Outside China (Excluding Hong Kong)')
ax.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for Profit Margin across regions
regions = ['North', 'South', 'East', 'West']
profits = [15000, 18000, 12000, 20000]
sales = [100000, 150000, 80000, 110000]

profit_margin = [profit / sale * 100 for profit, sale in zip(profits, sales)]

# Data for Inventory Levels
categories = ['Electronics', 'Clothing', 'Home Goods',"Beauty Products"]
average_inventory = [750, 1200, 500,300 ]
target_inventory = [500, 1500, 300,250]

# Visualizing Profit Margin
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Pie chart for Profit Margin
axs[0].pie(profit_margin, labels=regions, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'lightcoral', 'gold'])
axs[0].set_title('Profit Margin Across Regions')

# Bar chart for Inventory Levels
index = np.arange(len(categories))
bar_width = 0.35

bar1 = axs[1].bar(index, average_inventory, bar_width, label='Average Inventory')
bar2 = axs[1].bar(index + bar_width, target_inventory, bar_width, label='Target Inventory')

axs[1].set_xlabel('Product Category')
axs[1].set_ylabel('Inventory Level')
axs[1].set_title('Comparison of Average and Target Inventory Levels')
axs[1].set_xticks(index + bar_width / 2)
axs[1].set_xticklabels(categories)
axs[1].legend()

plt.tight_layout()
plt.show()
```
